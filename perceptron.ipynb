```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv('dataset_Perceptron.txt')

m = np.size(data,0)
n = np.size(data,1)

array = data.values

X = array[:,0:n-1]
Y = array[:,n-1]
```

```python
#Estandarizacion

media = X.mean(axis=0)
sigma = X.std(axis=0,ddof=1)

X = (X - media)/sigma


clase0 = np.argwhere(Y==0)
clase0 = clase0[:,0]

clase1 = np.argwhere(Y==1)
clase1 = clase1[:,0]
plt.plot(X[clase0,0],X[clase0,1],'cp',mec="black")
plt.plot(X[clase1,0],X[clase1,1],'mo',mec="black")

#Entreanmiento

modelo = linear_model.Perceptron()

modelo.fit(X,Y)

b = modelo.intercept_
W = modelo.coef_ 

x1 = np.linspace(-2,2,10)
x2 = (-b - x1*W[0,0]) / W[0,1] 


plt.plot(x1,x2,'-','r')
```

```python
# Hacer el dato de prueba 

datoPrueba = [[65,70]]

datoPrueba =  (datoPrueba - media)/sigma  

clasificacion = modelo.predict(datoPrueba)

print("PREDICCION :",clasificacion)

```
